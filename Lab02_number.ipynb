## B. C 언어에서의 데이터 타입과 표현 범위 알아보기

1. char, short, int, float, double 타입에 메모리 크기에 대해서 살펴보자.

2. 수의 표현 범위에 대해서 알아보자.

  - n비트 정수형 변수의 수의 표현범위
  - 양수의 경우: 0 - 2^n -1
  - 2의 보수로 표현한 음수의 경우: -2^(n-1) - 2^(n-1) -1
  - char 타입은?

3. 수의 최대/최소 표현 값을 알아보자.


````python
%%writefile mycode2.c

#include <stdio.h> // 헤더 화일 --> printf 함수를 사용하기 위해서 필요!

void main(void)
{
  char c1;
  short s1;
  int i1;
  float f1;
  double d1;

  // sizeof 함수: 주어진 변수가 몇 바이트를 할당 하는지 반환합니다!!!

  printf("문자 타입 char는 %d 바이트를 사용합니다.\n", (int) sizeof(c1));
  printf("짧은 정수 타입 short는 %d 바이트를 사용합니다.\n", (int) sizeof(s1));
  printf("정수 타입 int는 %d 바이트를 사용합니다.\n", (int) sizeof(i1));
  printf("float 타입 float는 %d 바이트를 사용합니다.\n", (int) sizeof(f1));
  printf("double 타입 double는 %d 바이트를 사용합니다.\n", (int) sizeof(d1));
}
````

````python
!gcc -o mycode2 mycode2.c
````

````python
!./mycode2
````

## C. 진수 변환을 해보자!

1. 16진수 표현
  - ```int a = 0x11```
  - 16진수로 프린트 하려면 포맷 ```%x``` 를 사용한다.

2. 8진수 표현
  - ```int a = O11```
  - 8진수로 프린트 하려면 포맷 ```%o``` 를 사용한다.

3. 2진수 표현 ???

````python
#진수 변환
%%writefile mycode5.c

#include <stdio.h>

void main(void)
{
  int a = 0x22;  // 2*16 + 2 = 34
  int b = 022;   // 2*8 + 2 = 18

  printf("a 변수의 값은 십진수로 %d 이고 16진수로 %x \n", a, a);
  printf("b 변수의 값은 십진수로 %d 이고 8진수로 %o \n", b, b);
}
````

````python
!gcc -o mycode5 mycode5.c
````

````python
!./mycode5
````

````python
# prompt: make a C code for printing binary value for a given integer.

%%writefile mycode6.c

#include <stdio.h>

void main(void)
{
  char a = 6;
  int i;
  for (i = sizeof(char) * 8; i >= 0 ; i--)
  {
    printf("%d", (a >> i) & 1);
  }
  printf("\n");
}
````

````python
!gcc -o mycode6 mycode6.c
````

````python
!./mycode6
````
